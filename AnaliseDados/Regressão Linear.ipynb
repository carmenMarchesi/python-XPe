#### Importando a biblioteca para visualização de dados

```python
import matplotlib.pyplot as plt
```

## Dados

```python
x = [-1., -0.77777778, -0.55555556, -0.33333333, -0.11111111, 0.11111111, 0.33333333, 0.55555556, 0.77777778, 1.]
y = [-1.13956201, -0.57177999, -0.21697033, 0.5425699, 0.49406657, 1.14972239, 1.64228553, 2.1749824, 2.64773614, 2.95684202]
```

## Plot dos Dados

```python
plt.figure(figsize=(10,5)) #Tamanho da figura 10x5
plt.plot(x, y, 'o', label='Dados Originais') # 'o' indica que quero um plot do tipo escater (com os pontos), 
# caso eu deixasse sem nada ele me traria um plot com uma linha
plt.legend() # expor o label
plt.xlabel("x") # Legenda do eixo x
plt.ylabel("y") # legenda do eixo y
plt.grid() # Mostra o grid ao fundo, de distância entre os valores
plt.show()
```

---------------------------  
Iremos estimar uma função do tipo: y = a*x + b, ou seja, devemos achar  
quais os valores de a e b que melhor representam os dados  
  
Os valores de a e b são: (a, b): 2, 1

## Transformando para numpy e vetor coluna

```python
x, y = np.array(x).reshape(-1, 1), np.array(y).reshape(-1, 1)
```

## Adicionando bias: para estimar o termo b

```python
X = np.hstack((x, np.ones(x.shape)))
```

```python
print(X)
```

## Estimulando a e b

```python
beta = np.linalg.pinv(X).dot(y)
print("a estimado:", beta[0][0])
print("b estimado:", beta[1][0])
```

## Plot dos dados

```python
plt.figure(figsize=(10,5)) 
plt.plot(x, y, 'o', label='Dados Originais') 
plt.plot(x, X.dot(beta), label='Regressão Linear') #x é o valor original do x e no y vamos
# plotar a estimativa do y, que é o x multiplicado matricialmente pelo beta.
# sem o 'o' ele traça uma linha.
plt.legend() 
plt.xlabel("x") 
plt.ylabel("y") 
plt.title("Regressão linear com numpy")
plt.grid() 
plt.show()
```
