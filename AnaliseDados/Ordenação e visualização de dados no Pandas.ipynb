# Ordenação no Pandas

```python
import pandas as pd
```

```python
df = pd.read_csv("https://pycourse.s3.amazonaws.com/temperature.csv")
df
```

```python
df['date'] = pd.to_datetime(df['date'])
df.dtypes
```

```python
df = df.set_index('date')
```

## Método sort_values e Método sort_index

### Ordenação crescente por uma coluna

```python
df.sort_values(by='temperatura')
```

```python
df.sort_values(by='classification')
```

### Ordenação crescente por mais de uma coluna

```python
df.sort_values(by=['classification', 'temperatura'])
```

### Ordenação decrescente por uma coluna

```python
df.sort_values(by='temperatura', ascending=False)
```

### Ordenação crescente pelo índice

```python
df.sort_index()
```

### Ordenação crescente pelo índice

```python
df.sort_index(ascending=False)
```

## Gráfico de setores (plot de "pizza")

**figsize** - Tamanho do plot  
**shadow** - Sombreamento  
**autopct** - Coloca as porcentagens - Com apenas 1 casa decimal para representar a porcentagem  
</br>

```python
df['classification'].value_counts().plot.pie(autopct='%1.1f%%', shadow=True, figsize=(10,7));
```
